```python
import os
import matplotlib.pyplot as plt
from PIL import Image

train_path = "dataset/train"

# Visualizar primeros 9 perros
dog_imgs = [os.path.join(train_path, "dogs", img) 
            for img in os.listdir(os.path.join(train_path, "dogs"))[:9]]

plt.figure(figsize=(8,8))
for i, img_path in enumerate(dog_imgs):
    img = Image.open(img_path)
    plt.subplot(3,3,i+1)
    plt.imshow(img)
    plt.axis("off")
plt.suptitle("Ejemplos de PERROS")
plt.show()

# Visualizar primeros 9 gatos
cat_imgs = [os.path.join(train_path, "cats", img) 
            for img in os.listdir(os.path.join(train_path, "cats"))[:9]]

plt.figure(figsize=(8,8))
for i, img_path in enumerate(cat_imgs):
    img = Image.open(img_path)
    plt.subplot(3,3,i+1)
    plt.imshow(img)
    plt.axis("off")
plt.suptitle("Ejemplos de GATOS")
plt.show()
```
